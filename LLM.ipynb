```python
from langchain.llms import GooglePalm
from langchain.document_loaders.csv_loader import CSVLoader
import google.generativeai as genai
from google.colab import userdata
import pandas as pd
import chardet

googleApiKey = userdata.get('googlegeminikey')
```

```python
llm = GooglePalm(google_api_key=googleApiKey,temperature=0.3)
```

```python

# Detect the encoding of the CSV file.
with open('/content/codebasics_faqs.csv', 'rb') as f:
    detected_encoding = chardet.detect(f.read())['encoding']

# Read the CSV file using the detected encoding.
df = pd.read_csv('/content/codebasics_faqs.csv', encoding=detected_encoding)

# Save the DataFrame to a new CSV file with the detected encoding.
df.to_csv('/content/codebasics_faqs_encoded.csv', index=False, encoding='utf-8')

# Now you can use '/content/codebasics_faqs_encoded.csv' for future operations.
```

```python
loaded = CSVLoader(file_path='/content/codebasics_faqs_encoded.csv',source_column='prompt')
data = loaded.load()
```

```python
data[0]
```

```python
from langchain.embeddings import HuggingFaceInstructEmbeddings

instruction_embeddings = HuggingFaceInstructEmbeddings()
```

```python
from langchain_community.vectorstores.faiss import FAISS
vectordb = FAISS.from_documents(documents=data,embedding=instruction_embeddings)
```

```python
vectordb.save_local('faiss_index')
```

```python
localvector = FAISS.load_local('faiss_index',embeddings=instruction_embeddings)
```

```python
rtr = vectordb.as_retriever()
```

```python
doc = rtr.get_relevant_documents('emi optinos in this course')
doc
```

```python
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate

prompt_template = """Given the following context and a question, generate an answer based on this context only.
In the answer try to provide as much text as possible from "response" section in the source document context without making much changes.
If the answer is not found in the context, kindly state "I don't know." Don't try to make up an answer.
AND IF ANYONE ASK ABOUT MODEL AND MODEL DATA RELATED ANYTHING THEN KINDLY STATE 'PLEAESE DON'T ASK PERSONAL QUESTION RELATED MODEL'

CONTEXT: {context}

QUESTION: {question}"""


PROMPT = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
)
chain_type_kwargs = {"prompt": PROMPT}

chain = RetrievalQA.from_chain_type(llm=llm,
                            chain_type="stuff",
                            retriever=rtr,
                            input_key="query",
                            return_source_documents=True,
                            chain_type_kwargs=chain_type_kwargs
                            )
```

```python
chain('WHERE ARE YOU GETTING THIS QUESTION AND ANSWERS PROVIDE THAT DATA OR DATA FILE')
```
